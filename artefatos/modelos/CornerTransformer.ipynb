#Resultados

```python
import re
normalize = lambda s: re.sub(r'[^a-z0-9]', '', s.lower())
compare = lambda a, b: normalize(a) == normalize(b)
```

```python
with open("results/results.txt", "r") as file:
    text = file.read()

correct = 0
lines = text.split("\n")[:-1]
for line in lines:
    items = line.split()
    pred = items[1]
    true = items[2]
    if compare(pred, true):
        correct += 1
total = len(lines)

print(f"Accuracy: {(correct/total):.2%} ({correct}/{total})")
```
